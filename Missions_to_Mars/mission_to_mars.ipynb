```python
#dependencies
from bs4 import BeautifulSoup as bs
import requests
import os
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
#url to be scrapped
scrap_url = 'https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest'
```

```python
# Retrieve page with the requests module
response = requests.get(scrap_url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
scrap_soup = bs(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(scrap_soup.prettify())
```

```python
#find title and paragraph of news
news_title = scrap_soup.find_all('div', class_='content_title')[0].text
news_p = scrap_soup.find_all('div', class_='article_teaser_body')

# print info 
print(news_title)
print(news_p)
```

```python
# find image url using splitter
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# find image url using splitter
#first define url
site_url = 'https://data-class-jpl-space.s3.amazonaws.com/'
pic_url = 'https://data-class-jpl-space.s3.amazonaws.com/JPL_Space/index.html'
browser.visit(pic_url)
html = browser.html

images_soup = bs(html, 'html.parser')
```

```python
#featured image link
image_path = images_soup.find_all('img')[1]["src"]
featured_image_url = site_url + image_path
print(featured_image_url)
```

```python
#mars hemispheres website through splinter 
hemispheres_url = "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"
browser.visit(hemispheres_url)
```

```python
# html object
html_hemispheres = browser.html

# parse
soup = bs(html_hemispheres, 'html.parser')

# get mars hemispheres info
items = soup.find_all('div', class_='item')

# make list for urls 
hemisphere_image_urls = []

# main_ul 
hemispheres_main_url = 'https://astrogeology.usgs.gov'

# start loop
for i in items: 
    # title
    title = i.find('h3').text
    
    # store link of full image website
    partial_img_url = i.find('a', class_='itemLink product-item')['href']
    
    # got to like with the full image 
    browser.visit(hemispheres_main_url + partial_img_url)
    
    # html obj of site
    partial_img_html = browser.html
    
    # parse html with every site
    soup = bs( partial_img_html, 'html.parser')
    
    # get img
    img_url = hemispheres_main_url + soup.find('img', class_='wide-image')['src']
    
    # append img_url and title.
    hemisphere_image_urls.append({"title" : title, "img_url" : img_url})
    


```
